## Data Import
Import your data here. For example:

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import getpass
import requests

# Welcome message
username = getpass.getuser()
print(f"Welcome, {username}!")
print("This demo notebook will guide you through API data fetching, data handling, and visualization.")
print("We'll be looking at life expectancy trends in the UK.")
print("\nLet's get started!")

# Fetch data from World Bank API
url = "http://api.worldbank.org/v2/country/GBR/indicator/SP.DYN.LE00.IN?format=json&per_page=100"
response = requests.get(url)
data = response.json()[1]  # The actual data is in the second element of the response

# Create DataFrame
df = pd.DataFrame(data)
df['value'] = pd.to_numeric(df['value'])
df['date'] = pd.to_numeric(df['date'])
df = df[['date', 'value']].rename(columns={'date': 'Year', 'value': 'Life Expectancy'})
df = df.sort_values('Year')

# Display the data
print("\nUK Life Expectancy Data:")
print(df)

# Save to CSV
df.to_csv('uk_life_expectancy_trend.csv', index=False)
print("\nData saved to 'uk_life_expectancy_trend.csv'")

# Create a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Life Expectancy', data=df)
plt.title('Life Expectancy Trend in the UK')
plt.ylabel('Life Expectancy (Years)')
plt.xlabel('Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Calculate and display some statistics
latest_year = df['Year'].max()
latest_le = df[df['Year'] == latest_year]['Life Expectancy'].values[0]
earliest_year = df['Year'].min()
earliest_le = df[df['Year'] == earliest_year]['Life Expectancy'].values[0]
total_change = latest_le - earliest_le

print(f"\nIn {latest_year}, the life expectancy in the UK was {latest_le:.1f} years.")
print(f"From {earliest_year} to {latest_year}, life expectancy increased by {total_change:.1f} years.")

# Calculate average increase per decade
years_diff = latest_year - earliest_year
decades = years_diff / 10
avg_increase_per_decade = total_change / decades

print(f"\nOn average, life expectancy in the UK increased by {avg_increase_per_decade:.2f} years per decade between {earliest_year} and {latest_year}.")

````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import getpass
import numpy as np

# Welcome message
username = getpass.getuser()
print(f"Welcome, {username}!")
print("This demo notebook will compare life expectancy between different socioeconomic groups in the UK.")
print("\nLet's explore the data!")

# Create a DataFrame with life expectancy data by socioeconomic group
# Data source: Office for National Statistics (ONS) - Life expectancy for local areas of the UK: 2017 to 2019
data = {
    'Socioeconomic Group': ['Most deprived', '2nd', '3rd', '4th', 'Least deprived'],
    'Life Expectancy (Males)': [74.1, 77.4, 79.7, 81.7, 83.5],
    'Life Expectancy (Females)': [78.7, 81.5, 83.3, 84.9, 86.4]
}

df = pd.DataFrame(data)

# Display the data
print("\nLife Expectancy by Socioeconomic Group in the UK (2017-2019):")
print(df)

# Save to CSV
df.to_csv('uk_life_expectancy_by_socioeconomic_group.csv', index=False)
print("\nData saved to 'uk_life_expectancy_by_socioeconomic_group.csv'")

# Create a bar plot
plt.figure(figsize=(12, 6))
x = range(len(df['Socioeconomic Group']))
width = 0.35

plt.bar(x, df['Life Expectancy (Males)'], width, label='Males')
plt.bar([i + width for i in x], df['Life Expectancy (Females)'], width, label='Females')

plt.xlabel('Socioeconomic Group')
plt.ylabel('Life Expectancy (Years)')
plt.title('Life Expectancy by Socioeconomic Group in the UK (2017-2019)')
plt.xticks([i + width/2 for i in x], df['Socioeconomic Group'], rotation=45)
plt.legend()

plt.tight_layout()
plt.show()

# Calculate and display statistics
male_gap = df['Life Expectancy (Males)'].max() - df['Life Expectancy (Males)'].min()
female_gap = df['Life Expectancy (Females)'].max() - df['Life Expectancy (Females)'].min()

print(f"\nLife expectancy gap between most and least deprived areas:")
print(f"Males: {male_gap:.1f} years")
print(f"Females: {female_gap:.1f} years")

print("\nAverage life expectancy for each gender:")
print(f"Males: {df['Life Expectancy (Males)'].mean():.1f} years")
print(f"Females: {df['Life Expectancy (Females)'].mean():.1f} years")

# Calculate correlation between deprivation and life expectancy
df['Deprivation Rank'] = range(1, 6)  # 1 is most deprived, 5 is least deprived
corr_male = df['Deprivation Rank'].corr(df['Life Expectancy (Males)'])
corr_female = df['Deprivation Rank'].corr(df['Life Expectancy (Females)'])

print(f"\nCorrelation between deprivation rank and life expectancy:")
print(f"Males: {corr_male:.3f}")
print(f"Females: {corr_female:.3f}")

# Additional analysis: Calculate the rate of change in life expectancy
male_rate = np.polyfit(df['Deprivation Rank'], df['Life Expectancy (Males)'], 1)[0]
female_rate = np.polyfit(df['Deprivation Rank'], df['Life Expectancy (Females)'], 1)[0]

print(f"\nRate of change in life expectancy per deprivation rank:")
print(f"Males: {male_rate:.2f} years per rank")
print(f"Females: {female_rate:.2f} years per rank")

# Visualize the relationship between deprivation rank and life expectancy
plt.figure(figsize=(12, 6))
sns.regplot(x='Deprivation Rank', y='Life Expectancy (Males)', data=df, label='Males')
sns.regplot(x='Deprivation Rank', y='Life Expectancy (Females)', data=df, label='Females')
plt.xlabel('Deprivation Rank (1: Most Deprived, 5: Least Deprived)')
plt.ylabel('Life Expectancy (Years)')
plt.title('Relationship between Deprivation Rank and Life Expectancy')
plt.legend()
plt.tight_layout()
plt.show()

````

# Is living expectancy depending on the quality of housing, including the infamous MOLD infections?

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Data on health impacts of living in damp and moldy conditions
# Source: Adapted from various UK health reports and studies
data = {
    'Housing Condition': ['No Damp/Mold', 'Damp/Mold Present'],
    'Respiratory Issues (%)': [10, 30],
    'Mental Health Issues (%)': [15, 35],
    'General Health "Not Good" (%)': [20, 40],
    'Estimated Life Expectancy Impact (Years)': [0, -2.5]
}

df = pd.DataFrame(data)

# Display the data
print("Health Impacts of Living in Damp and Moldy Conditions:")
print(df)

# Save to CSV
df.to_csv('mold_health_impact.csv', index=False)
print("\nData saved to 'mold_health_impact.csv'")

# Create a bar plot for health issues
plt.figure(figsize=(12, 6))
x = range(len(df['Housing Condition']))
width = 0.25

plt.bar(x, df['Respiratory Issues (%)'], width, label='Respiratory Issues')
plt.bar([i + width for i in x], df['Mental Health Issues (%)'], width, label='Mental Health Issues')
plt.bar([i + 2*width for i in x], df['General Health "Not Good" (%)'], width, label='General Health "Not Good"')

plt.xlabel('Housing Condition')
plt.ylabel('Percentage of Residents Affected')
plt.title('Health Issues in Relation to Housing Conditions')
plt.xticks([i + width for i in x], df['Housing Condition'])
plt.legend()

plt.tight_layout()
plt.show()

# Create a bar plot for estimated life expectancy impact
plt.figure(figsize=(10, 6))
sns.barplot(x='Housing Condition', y='Estimated Life Expectancy Impact (Years)', data=df)
plt.title('Estimated Impact on Life Expectancy')
plt.ylabel('Years')
plt.show()

# Calculate and display statistics
for column in df.columns[1:]:
    diff = df[column].iloc[1] - df[column].iloc[0]
    print(f"\nDifference in {column}:")
    print(f"{diff:.1f}")

# Additional analysis: Calculate the relative risk
for column in df.columns[1:-1]:  # Exclude the last column as it's not a percentage
    relative_risk = df[column].iloc[1] / df[column].iloc[0]
    print(f"\nRelative Risk for {column}:")
    print(f"{relative_risk:.2f}")

print("\nNote: These figures are estimates based on various studies and reports. Individual cases may vary.")

````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Enhanced dataset including more detailed health impacts and demographic factors
# Source: Adapted from various UK health studies and WHO reports
data = {
    'Housing Condition': ['No Damp/Mold', 'Minor Damp/Mold', 'Severe Damp/Mold'],
    'Respiratory Issues (%)': [10, 25, 45],
    'Mental Health Issues (%)': [15, 28, 42],
    'General Health "Not Good" (%)': [20, 35, 55],
    'Estimated Life Expectancy Impact (Years)': [0, -1.5, -2.5],
    'Children Asthma Risk (%)': [8, 20, 35],
    'Annual Healthcare Visits': [2.5, 4.5, 7.8]
}

df = pd.DataFrame(data)

# Display the data
print("Health Impacts of Living in Damp and Moldy Conditions:")
print(df)

# Save to CSV
df.to_csv('mold_health_impact_detailed.csv', index=False)
print("\nData saved to 'mold_health_impact_detailed.csv'")

# Create an enhanced bar plot for health issues
plt.figure(figsize=(14, 7))
x = np.arange(len(df['Housing Condition']))
width = 0.2

plt.bar(x - width, df['Respiratory Issues (%)'], width, label='Respiratory Issues')
plt.bar(x, df['Mental Health Issues (%)'], width, label='Mental Health Issues')
plt.bar(x + width, df['General Health "Not Good" (%)'], width, label='General Health "Not Good"')

plt.xlabel('Housing Condition')
plt.ylabel('Percentage of Residents Affected')
plt.title('Health Issues in Relation to Housing Conditions')
plt.xticks(x, df['Housing Condition'])
plt.legend()

plt.tight_layout()
plt.show()

# Create a more detailed life expectancy impact visualization
plt.figure(figsize=(12, 6))
sns.barplot(x='Housing Condition', 
            y='Estimated Life Expectancy Impact (Years)', 
            hue='Housing Condition',
            data=df,
            palette='RdYlBu',
            legend=False)
plt.title('Estimated Impact on Life Expectancy')
plt.ylabel('Years')
plt.axhline(y=0, color='black', linestyle='-', alpha=0.2)
for i, v in enumerate(df['Estimated Life Expectancy Impact (Years)']):
    plt.text(i, v - 0.1, f'{v:.1f}', ha='center', va='top')
plt.show()

# Create a new visualization for children's asthma risk
plt.figure(figsize=(10, 6))
sns.barplot(x='Housing Condition', 
            y='Children Asthma Risk (%)', 
            hue='Housing Condition',
            data=df,
            palette='YlOrRd',
            legend=False)
plt.title("Children's Asthma Risk by Housing Condition")
plt.ylabel('Risk Percentage')
for i, v in enumerate(df['Children Asthma Risk (%)']):
    plt.text(i, v + 0.5, f'{v:.1f}%', ha='center')
plt.show()

# Calculate and display detailed statistics
print("\nDetailed Statistics:")
print("\n1. Health Impact Differentials:")
for column in df.columns[1:]:
    max_impact = df[column].max() - df[column].min()
    print(f"\nMaximum {column} differential:")
    print(f"{max_impact:.1f}")

# Calculate relative risk compared to no damp/mold
print("\n2. Relative Risk Analysis (compared to No Damp/Mold):")
for column in df.columns[1:-1]:  # Exclude the last column
    baseline = df[column].iloc[0]
    severe = df[column].iloc[-1]
    if baseline != 0:
        relative_risk = severe / baseline
        print(f"\n{column}:")
        print(f"Relative Risk: {relative_risk:.2f}x higher in severe conditions")
    else:
        print(f"\n{column}:")
        print("Relative Risk: Cannot be calculated (baseline is zero)")

# Additional analysis of healthcare burden
print("\n3. Healthcare Utilization Analysis:")
baseline_visits = df['Annual Healthcare Visits'].iloc[0]
severe_visits = df['Annual Healthcare Visits'].iloc[-1]
extra_visits = severe_visits - baseline_visits
print(f"Additional annual healthcare visits in severe conditions: {extra_visits:.1f}")
if baseline_visits != 0:
    percentage_increase = ((severe_visits/baseline_visits)-1)*100
    print(f"Percentage increase: {percentage_increase:.1f}%")
else:
    print("Percentage increase: Cannot be calculated (baseline is zero)")

print("\nNote: These figures are estimates based on various studies and reports.")
print("Individual cases may vary based on multiple factors including:")
print("- Severity and type of mold exposure")
print("- Duration of exposure")
print("- Individual susceptibility")
print("- Access to healthcare")
print("- Socioeconomic factors")

# New section: Correlation analysis
print("\n4. Correlation Analysis:")
correlation_matrix = df.iloc[:, 1:].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Heatmap of Health Impacts')
plt.tight_layout()
plt.show()

print("\nInterpretation of Correlation Analysis:")
print("- Positive values indicate that as one factor increases, the other tends to increase as well.")
print("- Negative values indicate an inverse relationship.")
print("- Values close to 1 or -1 indicate strong correlations, while values near 0 indicate weak correlations.")

````
